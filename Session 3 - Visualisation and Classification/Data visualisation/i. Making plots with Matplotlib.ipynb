# Line plots

**Purpose**: Showing trends in data -- usually time series data with many time points.

**matplotlib function**: [`plot(x, y)`](http://matplotlib.org/api/pyplot_api.html#matplotlib.pyplot.plot)

* `x`: The x-coordinates of the lines or markers.

* `y`: The y-coordinates of the lines or markers.

Let's start in the simplest way possible and plot $y=x^2$ in a line graph

```python
x = np.array([0, 1, 2,3,4])
y=np.exp(x)
plt.plot(x, y, '-', linewidth=3)
;
```

Great, that was easy. That was probably the way you've done it before. However, that style of plotting stuff does not follow the typical 'matplotlib workflow'. Let me tell you a bit more about it and why it's good to keep that structure when plotting. 

In order to learn how to properly create and tweak graphs, let's break it down in a few steps. There is a matplotlib framework that allows you to create advanced graphs with just a few commands, these steps are:

1. Start by creating a figure. The figure is basically the axes that are going to contain you data series to be plotted in there. The command to create a new figure in your notebook is **plt.figure()**. You can define figure size here. You will see later on how you can create multiple graphs in one figure. That's why it's good practice to start by defining the figure first, rather than going straight in for the plot. 

2. Plot the data. There can be multiple data series in one plot, for example **plt.plot('x')** and plt.plot('y') will give a line plot of both x and y in the same plot in the figure. (mind you there can also be multiple sub-plots in one figure, each containing multiple data series) You can also do plt.bar(x), to get a different type of plot. The term "plt.plot()" is reserved for normal line plots. 

3. Configure the axes, you might want to add a series as tickmarks for one of the axes. **plt.xticks()** and plt.yticks() allow you to do this, and **plt.xlim()** and plt.ylim() set the range of your axes. 

4. Add annotation. You should consider adding labels and legends to the data series plotted, and a graph title. You can use **plt.xlabel()**, plt.ylabel(), **plt.legend()**, **plt.title()**, and **plt.grid()** to show gridlines etc 

5. Show or save the figure to display the result. You can use **plt.show()** just to show, we won't need this because of our settings above (%matplotlib inline), which causes the figures to be shown within the notebook. **plt.savefig()** allow you to store the figure to a designated folder

So now that we have a framework for using matplotlib, let's give it some practice! Let's return to creating line graphs

```python
# np.linspace returns num evenly spaced numbers over the specified interval 0-20
x = np.linspace(0, 20)

#numpy contains functions such as sin, cos and exp which are called by np.sin, np.cos etc
y1 = np.sin(x)
y2 = np.sin(x - np.pi)

#Start the figure
plt.figure()

#add two lines
plt.plot(x, y1)
plt.plot(x, y2)

# and done, let's keep it simple
# remember we should use plt.show() but we don't need it because of our settings!
;
```

With the plt.plot() command there are a few parameters we can tweak. This is a few of them:

**Commonly used parameters**:

* `color`: Set the color of the line.

* `linestyle`: Set the line style, e.g., solid, dashed, or none.

* `linewidth`: Set the line thickness.

* `marker`: Set the marker style, e.g., circles, triangles, or none.

* `markersize`: Set the marker size.

* `label`: Set the label for the line that will show up in the legend.

```python
#this is our graph data
x = np.linspace(0, 20)
y1 = np.sin(x)
y2 = np.sin(x - np.pi)

# start the figure
plt.figure()

#set a bunch of parameters under the plt.plot() commands
plt.plot(x,
         y1,
         color='black',
         linestyle='-',
         linewidth=2,
         marker='s',
         markersize=6,
         label='y1')

plt.plot(x,
         y2,
         color='gray',
         linestyle='--',
         linewidth=2,
         marker='^',
         markersize=6,
         label='y2')

# and let's add a legend
plt.legend()
# and done
;
```

# Vertical bar charts

Let's move on. We've seen line charts, let's now get real crazy and plot some bar charts. We've actually already seen these a few times in our previous exercises, but let's have a deeper look.

**Purpose**: Comparing categories **OR** showing temporal trends in data with few (< 4) time points.

**matplotlib function**: [`bar(bottom, height)`](http://matplotlib.org/api/pyplot_api.html#matplotlib.pyplot.bar)

* `bottom`: The x coordinate(s) on the bottom of the bars.

* `height`: The height(s) of the bars.

Alright here we go:

```python
years = np.arange(2012, 2015)
values = [2, 5, 9]

plt.figure()

plt.bar(years, values)
;
```

# Horizontal bar charts

**Purpose**: Comparing categories.

**matplotlib function**: [`barh(left, width)`](http://matplotlib.org/api/pyplot_api.html#matplotlib.pyplot.barh)

* `width`: The width(s) of the bars.

Let's give it  a try:

```python
categories = ['A', 'B', 'C', 'D', 'E']
values = [7, 12, 4, 2, 9]

plt.figure()

plt.barh(categories, values)
```

Ok, that's slightly better but still useless. Now I've converted those categories into numerical values, but I don't know what corresponds to what. We need to add some yticks to this to know what's going on.

# Pie charts

**Purpose**: Displaying a simple proportion.

**matplotlib function**: [`pie(sizes)`](http://matplotlib.org/api/pyplot_api.html#matplotlib.pyplot.pie)

* `sizes`: The size of the wedges as either a fraction or number.

Let's start off very simple:

```python
counts = [17, 14]

plt.figure()

plt.pie(counts)
;
```

Wait...what? Oval pies? That's weird. 

Well this is because you haven't defined the figure-size to be square, so let's do that. We can give plt.figure() a size parameter, and when it's a square, you're pie will look normal. 

```python
counts = [17, 14]

plt.figure(figsize=(4, 4))

plt.pie(counts)
;
```

Much better, let's have a look at what we can tweak. 

**Commonly used parameters**:

* `colors`: Set the colors of the wedges.

* `labels`: Set the labels of the wedges.

* `startangle`: Set the angle that the wedges start at.

* `autopct`: Set the percentage display format of the wedges.

```python
counts = [17, 14]

plt.figure(figsize=(4, 4))

plt.pie(counts,
        colors=['blue', 'orange'],
        labels=['Category A', 'Other categories'],
        startangle=90,
        autopct='%1.1f%%')
;
```

# Scatter plots

One of the most useful plots are perhaps scatter plots. When exploring your data, and seeing how your features relate to your target variable, you'll be using these a lot. 

**Purpose**: Displaying relationships between variables.

**matplotlib function**: [`scatter(x, y)`](http://matplotlib.org/api/pyplot_api.html#matplotlib.pyplot.scatter)

* `x`, `y`: The values for the two variables.

Let's look at an example

```python
x = range(20)
y = np.arange(50, 70) + (np.random.random(20) * 10.)

plt.figure()

plt.scatter(x, y)
;
```

We can also tweak this stuff. 

**Commonly used parameters**:

* `c`: Set the color of the markers.

* `s`: Set the size of the markers.

* `marker`: Set the marker style, e.g., circles, triangles, or squares.

* `edgecolor`: Set the color of the lines on the edges of the markers.

```python
x = np.arange(20)
y = np.arange(50, 70) + (np.random.random(20) * 10.)

plt.figure()

plt.scatter(x,
            y,
            c='red',
            s=40,
            marker='s',
            edgecolor='none')
;
```

The area of each point can actually be a very useful thing to play with. It can be a way to introduce another variable to the graph, by allowing the marker size to vary (e.g. big dot for high value on third feature, little dot for low value on that third feature. Let me give you an example on how to do this:

```python
N = 50
x = np.random.rand(N)
y = np.random.rand(N)
# area is the third variable in the graph, and varies from 0 to 15 point radiuses for the circle
area = np.pi *(15 * np.random.rand(N))**2 

plt.figure()
plt.scatter(x, y, s=area, alpha=0.5)
;
```

If your graph is looking very grainy, you can increase the resolution. The option needs to be given under the savefig command, and is as follows: plt.savefig('name'.png, **dpi=200**). # Subplots

There is a way to get multiple plots into one figure, which comes in handy when you have to create many different plots. You've already seen these types of plots before in our exercises, so let me show you how to do this.

**Purpose:** Allows you to place multiple charts in a figure.

**matplotlib function:** [``subplot(nrows, ncols, plot_number)``](http://matplotlib.org/api/pyplot_api.html#matplotlib.pyplot.subplot)

* `nrows`: The number of rows in the figure.

* `ncols`: The number of columns in the figure.

* `plot_number`: The placement of the chart (starts at 1, and not at 0 as our arrays, dataframes and series do, confusion alert!).

Let's get started. Remember I talked about tight_layout before? When you're working with many subplots you have to think about how to fit stuff into a page/one graph, so we're going to apply it here. 

```python
plt.figure()

#we need to start a little loop that's going to count from 1 up to but not including 7
for i in range(1, 7):
    #each number will be assigned a place in my 3 by 2 set of graphs 
    plt.subplot(3, 2, i)
    plt.title(i)
    plt.xticks([])
    plt.yticks([])

plt.tight_layout()
;
```

Subplots can take any form or shape, and they don't have to be the same as all the other subplots in the figure. I'll show you how to put 2 different types of plots into 1 figure. 

```python
dist1 = np.random.normal(42, 7, 1000)
dist2 = np.random.normal(59, 3, 1000)

plt.figure(figsize=(10, 4))

plt.subplot(1, 2, 1)
plt.hist(dist1)
plt.title('dist1')

plt.subplot(1, 2, 2)
plt.scatter(dist2, dist1)
plt.xlabel('dist2')
plt.ylabel('dist1')

plt.tight_layout()
;
```

As you can see subplots take the same arguments as normal plots do, you can add titles, axis-labels, legends, etc etc. 

Now let's think about how we can repeat the same plot across many different subplots. This is useful for comparing many different categories of the same thing. You could loop over a set of different values and re-create the same plot. Let's start with plotting a set of different time-series data over a year range. 

```python
years = np.arange(2010, 2016)

plt.figure()

for category_num in range(1, 17):
    y_vals = np.arange(10, 16) + (np.random.random(6) * category_num / 4.)
    plt.plot(years, y_vals)

# add range to our y-axis    
plt.ylim(10, 18)
# we've seen this before, add year labels on the x-axis, but use year as string, otherwise we get those funky values
plt.xticks(years, [str(year) for year in years])
;
```

From the code above, can you deduce how many lines we have plotted? 

To make sure we understand what the loop above does, it sometimes helps to print stuff. So let's print those different series of y_vals that were created in the loop. Have a look:

```python
for category_num in range(1, 17):
    print(np.arange(10, 16) + (np.random.random(6) * category_num / 4.))
```

Now let's repeat the many lines chart, but this time we're going to use subplots to show each line individually. Notice that this time the set-up for the graph is the same (i.e. each subplot is the same graph, same range, same labels etc) so we can put the creation of the graph inside the loop we used earlier. 

```python
years = np.arange(2010, 2016)

plt.figure(figsize=(10, 10))

#plt.ylim and plt.xticks etc go inside the loop, as we want those settings for each subplot
for category_num in range(1, 17):
    plt.subplot(4, 4, category_num)
    y_vals = np.arange(10, 16) + (np.random.random(6) * category_num / 4.)
    plt.plot(years, y_vals)
    plt.ylim(10, 18)
    plt.xticks(years, [str(year) for year in years])
    plt.title('Category {}'.format(category_num))
    
plt.tight_layout()
;
```

# Matplotlib style

There are many ways to change the look of your graphs. There are parameters to play with under the standard configuration of matplotlib, which you can change. You can have a look for example here: http://matplotlib.org/users/customizing.html#dynamic-rc-settings

It is however time consuming to manually set the style configuration whenever you're creating new plots. I would advise you to use some pre-set styles, that are commonly used. Let's first create some data we can graph:

```python
x1_values = [2012, 2013, 2014, 2015]
y1_values = [4.3, 2.5, 3.5, 4.5]

x2_values = [2012, 2013, 2014, 2015]
y2_values = [2.4, 4.4, 1.8, 2.8]

x3_values = [2012, 2013, 2014, 2015]
y3_values = [2, 2, 3, 5]
```

Without having changed anything about the style, let's just have a look at plot for this data 

```python
plt.figure()
plt.plot(x1_values, y1_values, label='London')
plt.plot(x2_values, y2_values, label='Hong Kong')
plt.plot(x3_values, y3_values, label='San Francisco')

plt.xlim(2012, 2015)
plt.ylim(0, 6)
plt.xticks([2012, 2013, 2014, 2015], ['2012', '2013', '2014', '2015'])
plt.yticks([1, 2, 3, 4, 5])

plt.xlabel('')
plt.ylabel('% increase housing prices')

plt.legend(loc='upper center', ncol=3)
plt.grid(True)

plt.savefig('price_graph.png', dpi=150)
```

OK not bad, but can we make it look better? Let me show you first which styles are available:

```python
plt.style.available
```

Wow that's a lot. Both gg-plot and the seaborn styles are used quite often actually. The grayscale is one to remember when you know you'll have to output something in black and white. Let's pick one style and set it to the graph similar to above. 

```python
with plt.style.context('seaborn-white'):
    plt.plot(x1_values, y1_values, label='London')
    plt.plot(x2_values, y2_values, label='Hong Kong')
    plt.plot(x3_values, y3_values, label='San Francisco')

    plt.xlim(2012, 2015)
    plt.ylim(0, 6)
    plt.xticks([2012, 2013, 2014, 2015], ['2012', '2013', '2014', '2015'])
    plt.yticks([1, 2, 3, 4, 5])

    plt.xlabel('')
    plt.ylabel('% increase housing prices')

    plt.legend(loc='upper center', ncol=3)
    plt.grid(True)

    plt.savefig('pretty_graph.png', dpi=150)
    
```
